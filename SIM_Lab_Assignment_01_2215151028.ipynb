```python
# Necessary Libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```



---



# Task 1: Generate two vectors with 15 random floats each and plot them

```python

# Task 1: Generate two vectors with 15 random floats each and plot them
x = np.random.rand(15)
y = np.random.rand(15)

plt.plot(range(1, 16), x, marker='o', linestyle='-', label='Vector X', color='b')
plt.plot(range(1, 16), y, marker='s', linestyle='--', label='Vector Y', color='g')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Graph of Two Random Vectors')
plt.legend()
plt.grid(True)
plt.show()


```

# Task 2: Create a 4x4 matrix with random values and visualize it as a heatmap


```python
# Task 2: Create a 4x4 matrix with random values and visualize it as a heatmap
matrix = np.random.rand(4, 4)
sns.heatmap(matrix, annot=True, cmap='Blues', linewidths=0.7)
plt.title('Heatmap of 4x4 Random Matrix')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()

```

# Task 3:  Generate two 4x4 matrices, perform operations, and visualize results

```python
# Task 3: Generate two 4x4 matrices, perform operations, and visualize results
A = np.random.randint(1, 11, (4, 4))
B = np.random.randint(1, 11, (4, 4))

matrix_sum = A + B
matrix_diff = A - B
matrix_prod = np.dot(A, B)
# Print matrices
print("\nMatrix A:\n", A, "\n")
print("Matrix B:\n", B, "\n")
print("Matrix Addition:\n", matrix_sum, "\n")
print("Matrix Subtraction:\n", matrix_diff, "\n")
print("Matrix Multiplication:\n", matrix_prod, "\n")

# Visualizing results using bar plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Matrix Addition
axes[0].bar(range(1, 17), matrix_sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Matrix Subtraction
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Matrix Multiplication
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
